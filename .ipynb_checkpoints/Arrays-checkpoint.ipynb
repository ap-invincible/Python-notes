```python
#Creating an int array using lists.
class arr:
    def __init__(self):
        self.array = []
    def append(self,element):
        if isinstance(element,int):
            self.array.append(element)
            return self.array
        else:
            return f"This array only supports 'int' values"
    def insert(self,index,item):
        if isinstance(item,int):
            self.array.insert(index,item)
            return self.array
        else:
            return "Unsupported data type"
    def delete(self,index):
        del self.array[index]
        return self.array
    def remove(self,item):
        self.array.remove(item)
        return self.array
    def show(self):
        return self.array

#Checking functionality
myarray = arr()
myarray.append(10)
myarray.append(19)
myarray.append(91)
myarray.append(13)
myarray.append(6)
myarray.append(7)
myarray.show()
myarray.delete(3)
myarray.insert(3,14)
```

#########

Bubble Sort

#########

The highest value bubbles up in this algorithm

```python
x1 = [19,73,1,92,3,84,73,45,12,21]

def bubble_sort(array):
    n = len(array)
    for x in range(n-1): #x = 0 for the first time the outer loop runs, hence no element is fixed the first time this loop runs. This loop starts fixing elements from the second pass.
        for y in range(n-x-1): #Range = Len(array) - 1 - x. Here x keeps incrementing everytime the outer loop runs.
            if array[y] > array[y+1]:
                array[y], array[y+1] = array[y+1], array[y]
    print(array)

bubble_sort(x1) 
```




HOW THE ABOVE BUBBLE SORT FUNCTION?
    1. Everytime the outer loop(pass) runs -> One element is pushed to its final correct position.
    2. 1st pass -> Largest element settles
       2nd pass -> 2nd Largest element settles
       so on..
    3. When (n-1) elements are fixed, the last one remaining is automatically at the correct position.
    4. Thats why only (n-1) passes are required to sort the array.
    
    
LOOP FUNCTIONING:
    1. When the x loop runs for first time -> x has a value of '0'.
    2. The y variable also contains 0 for now.
    3. When the x loop runs 2nd time, x = 1.

Selection Sort

##########

Opposite of Bubble Sort.
The lowest value is moved to the front of the list along with each pass until the array is completely sorted.


```python
x1 = [19,73,1,92,3,84,73,45,12,21]

def selection_sort(array):
    n = len(array)
    for i in range(n-1):
        min_index = i
        for k in range(i+1,n):
            if array[k] < array[min_index]:
                min_index = k
        min_value = array.pop(min_index)
        array.insert(i, min_value)
    print(array)

selection_sort(x1)
```

Insertion Sort

##########

This sort holds values in two blocks of a single array.
One block is sorted and the other one in unsorted.
One value from the unsorted block is inserted at the correct place in the sorted block of the array.

Step 1: We start with an unsorted array.
[ 7, 12, 9, 11, 3]

Step 2: We can consider the first value as the initial sorted part of the array. If it is just one value, it must be sorted, right?
[ 7, 12, 9, 11, 3]

Step 3: The next value 12 should now be moved into the correct position in the sorted part of the array. But 12 is higher than 7, so it is already in the correct position.
[ 7, 12, 9, 11, 3]

Step 4: Consider the next value 9.
[ 7, 12, 9, 11, 3]

Step 5: The value 9 must now be moved into the correct position inside the sorted part of the array, so we move 9 in between 7 and 12.
[ 7, 9, 12, 11, 3]

Step 6: The next value is 11.
[ 7, 9, 12, > 11, 3]

Step 7: We move it in between 9 and 12 in the sorted part of the array.
[ 7, 9, 11, 12, 3]

Step 8: The last value to insert into the correct position is 3.
[ 7, 9, 11, 12, 3]

Step 9: We insert 3 in front of all other values because it is the lowest value.
[ 3,7, 9, 11, 12]

Finally, the array is sorted.
