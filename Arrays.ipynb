```python
#Creating an int array using lists.
class arr:
    def __init__(self):
        self.array = []
    def append(self,element):
        if isinstance(element,int):
            self.array.append(element)
            return self.array
        else:
            return f"This array only supports 'int' values"
    def insert(self,index,item):
        if isinstance(item,int):
            self.array.insert(index,item)
            return self.array
        else:
            return "Unsupported data type"
    def delete(self,index):
        del self.array[index]
        return self.array
    def remove(self,item):
        self.array.remove(item)
        return self.array
    def show(self):
        return self.array

#Checking functionality
myarray = arr()
myarray.append(10)
myarray.append(19)
myarray.append(91)
myarray.append(13)
myarray.append(6)
myarray.append(7)
myarray.show()
myarray.delete(3)
myarray.insert(3,14)
```

Bubble Sort

#########

The highest value bubbles up in this algorithm

```python
x1 = [19,73,1,92,3,84,73,45,12,21]

def bubble_sort(array):
    n = len(array)
    for x in range(n-1): #x = 0 for the first time the outer loop runs, hence no element is fixed the first time this loop runs. This loop starts fixing elements from the second pass.
        for y in range(n-x-1): #Range = Len(array) - 1 - x. Here x keeps incrementing everytime the outer loop runs.
            if array[y] > array[y+1]:
                array[y], array[y+1] = array[y+1], array[y]
    print(array)

bubble_sort(x1) 
```




HOW THE ABOVE BUBBLE SORT FUNCTION?
    1. Everytime the outer loop(pass) runs -> One element is pushed to its final correct position.
    2. 1st pass -> Largest element settles
       2nd pass -> 2nd Largest element settles
       so on..
    3. When (n-1) elements are fixed, the last one remaining is automatically at the correct position.
    4. Thats why only (n-1) passes are required to sort the array.
