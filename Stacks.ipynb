Stacks This data structure in python allows storage of multiple elements like a list or an array. #The last element to be added is the first one to be removed.

Basic Stack operations: 1. Push - Add at the last 2. Pop - Remove the last element 3. Peek - Return the last element 4. isEmpty - Check if the stack is empty 5. Size - Return the stack size Stacks can be implemented using Arrays or linked lists.


```python
s1 = []

#Push
def s1_push(element):
    s1.append(element)
    print(s1)
s1_push(16)
s1_push(19)
s1_push(21)
s1_push(80)

#Peek
def s1_peek():
    top = s1[-1]
    print(top)
s1_peek()

#Pop
def s1_pop():
    del s1[-1]
    print(s1)
s1_pop()

#Size
def s1_size():
    print('Size of stack =' , len(s1))
s1_size()

#isEmpty - 2 ways : 1.if len = 0 | 2.not bool
#not bool 
def s1_isEmpty():
    temp = not bool(s1)
    print('Is the stack empty :', temp)
s1_isEmpty()

# len =0
def s1_isEmpty_alt():
    if len(s1) == 0:
        print('Stack isEmpty?:', True)
    else:
        print('Stack isEmpty?:', False)
s1_isEmpty_alt()
```

Creating a stack using class

####################

```python
class stack:
    def __init__(self):
        self.stack = []

    #isEmpty is defined first because it might get called in the other methods.
    def isEmpty(self):
        lenstack = len(self.stack)
        if lenstack == 0:
            return f"Stack Empty : {True}"
        else:
            return f"Stack Empty : {False}\n Size : {len(self.stack)}"

    #Pop method
    def pop(self):
        if self.isEmpty() == False:
            del self.stack[-1]
        else:
            return self.isEmpty()
        return self.stack.pop()

    #Push method
    def push(self, element):
        self.stack.append(element)

    #Peek method
    def peek(self):
        if self.isEmpty() == True:
            return self.isEmpty()
        else:
            return self.stack[-1]

    #Size Method
    def size(self):
        lenstack = len(self.stack)
        if lenstack > 0:
            return len(self.stack)
        else:
            return self.isEmpty()

    #Min Method
    def min(self):
        minval = self.stack[0]
        for k in self.stack:
            if k < minval:
                minval = k
        return f"Minimum Value in Stack : {minval}"

    #Max Method
    def max(self):
        maxval = self.stack[0]
        for k in self.stack:
            if k > maxval:
                maxval = k
        return f"Maximum Value in Stack : {maxval}"

#myStack acts as self.
myStack = stack()
#print(type(myStack))

print(myStack.isEmpty())
myStack.push(90)
myStack.push(30)
myStack.push(40)
myStack.push(79)
myStack.push(89)
myStack.push(50)
print(myStack.peek())
print(myStack.size())
print(myStack.isEmpty())
print(myStack.min())
print(myStack.max())
```
