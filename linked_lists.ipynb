```python
class node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

node1 = node(10)
node2 = node(20)
node3 = node(30)
node4 = node(40)

node1.next = node2
node2.next = node3
node3.next = node4

node4.prev = node3
node3.prev = node2
node2.prev = node1

currentnode = node1
while currentnode:
    print(currentnode.data , end = " ")
    currentnode = currentnode.next

nodex = node4
while nodex:
    print(nodex.data , end = " ")
    nodex = nodex.prev
```
